```python
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt #to display our wordcloud
from PIL import Image #to load our image
import numpy as np #to get the color of our image
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
factory = StopWordRemoverFactory()
stopwords = factory.get_stop_words()

#sumber data https://www.youtube.com/watch?v=XBI_0UWf9jc
text = open('dataset.txt', 'r').read()
stopwords = set(stopwords)
stopwords.update(["yang", "dan", "di", "juga", "yg", "ini","itu","ada","kalo","ya","dari","lu","aja","gak","lah","lagi","jadi"
                 ,"bisa","tapi","mau","sudah","sama","udah","tuh","gk","tau","apa","bukan","nya","iya","cara","buat"])
```

```python
#Appearance-related
custom_mask = np.array(Image.open('cloud.png')) 
wc = WordCloud(background_color = 'white',
               stopwords = stopwords,
               mask = custom_mask,
               contour_width = 3,
               contour_color = 'black')
```

```python
wc.generate(text)
image_colors = ImageColorGenerator(custom_mask)
wc.recolor(color_func = image_colors)
```

```python
#Plotting
plt.figure(figsize=[15,15])
plt.imshow(wc, interpolation = 'bilinear')
plt.axis('off')
plt.show()

#wc.to_file('hasil.png')
```
